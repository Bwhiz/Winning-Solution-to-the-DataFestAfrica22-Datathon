Extracting hour of day In other to drill down into recommendations for **TAT** ( Turn - Around - Time) :

```python
data_service['Hour_of_day'] = data_service['Ticket Open Time'].dt.hour
```

Feature to know the hours taken to resolve a ticket;

```python
data_service['resolution_hour'] = data_service['Resolution_time_diff'].dt.total_seconds()/3600
```

**formula** to replicate resolution matrix:

```python
def hour_conversion(row_value):
    if row_value < 0.5:
        return 'Excellent'
    elif row_value >= 0.5 and row_value <= 1:
        return 'Good'
    elif row_value > 1 and row_value <= 3:
        return 'Fair'
    elif row_value > 3:
        return 'Critical'
```

```python
hour_conversion(4.368056)
```

```python
data_service['Resolution_class'] = data_service['resolution_hour'].apply(hour_conversion)
```

```python
data_service['Resolution_class'].value_counts()
```

```python
data_service
```

feature to showing total response seconds for a report:

```python
data_service['total_response_seconds'] = data_service['Response_time_diff'].dt.total_seconds()
```

```python
data_service
```

##### Formula to classify response time <= 10secs as good and > 10secs as bad

```python
def convert_response_time(time_diff):
    if time_diff <= 10:
        return 'Good'
    elif time_diff > 10:
        return 'Bad'
```

creating a column based on the formula above:

```python
data_service['response_time_status'] = data_service['total_response_seconds'].apply(convert_response_time)
```

```python
data_service
```

```python
data_service['response_time_status'].value_counts()
```

#### Formula to easily filter tickets that were reported within 6-9pm and outside 6-9pm

```python
def hour_of_day(row):
    if row >= 18 and row <= 21:
        return 'Within 6-9pm'
    else:
        return 'outside 6-9pm'
```

```python
data_service['within_or_outside_6_9pm'] = data_service['Hour_of_day'].apply(hour_of_day)
```
